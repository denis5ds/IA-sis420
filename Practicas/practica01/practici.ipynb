```python
# librerias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Cargar los datos.
data = pd.read_csv('pesoyaltura.csv')

# calculo de la pendiente m y recta b.
x = data['Estatura (m)']
y = data['Peso (kg)']
m = np.sum((x - np.mean(x)) * (y - np.mean(y))) / np.sum((x - np.mean(x)) ** 2)
b = np.mean(y) - m * np.mean(x)

# Crear los valores de y basados en la fórmula de la recta
ylinea = m * x + b

# Visualización de los datos generados y la recta ajustada
plt.scatter(data['Estatura (m)'], data['Peso (kg)'], color='black', label='Datos')
plt.plot(x, ylinea, color='red', label=f'Línea ajustada: y = {m:.2f}x + {b:.2f}')
plt.xlabel('Estatura (m)')
plt.ylabel('Peso (kg)')
plt.legend()
plt.show()

# Imprimir los valores de la pendiente y la intersección
print(f"Pendiente (m): {m:.2f}")
print(f"Intersección (b): {b:.2f}")
```
